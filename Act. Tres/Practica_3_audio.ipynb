```python
from openai import OpenAI
client = OpenAI()

audio_file= open("", "rb")#ruta archivo de audio
transcription = client.audio.transcriptions.create(
  model="whisper-1",
  file=audio_file
)
print(transcription.text)
```

```python
audio_file= open("", "rb") #ruta archivo de audio
transcription = client.audio.transcriptions.create(
model="whisper-1",
file=audio_file,
language="es"
)
print(transcription.text)
```

```python
audio_file= open("", "rb") #ruta archivo de audio
translation = client.audio.translations.create(
model="whisper-1",
file=audio_file,
)
print(translation.text)
```
